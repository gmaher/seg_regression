```python
import numpy as np
import pandas as pd
import matplotlib

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
matplotlib.rc('text', usetex=True)
matplotlib.rcParams.update({"font.size":16})
import os
import sys

sys.path.append(os.path.abspath('../..'))
from modules import io

from imageio import imread
```

```python
MODEL = "coronary"

gen = 1

# Ws = 100
# We = 200
# Hs = 40
# He = 260

Ws = 200
We = 600
Hs = 180
He = 600

modes = 20

im_fn = '/exterior.png'

DIR = "/media/marsdenlab/Data1/UQ/"+MODEL+"/figures/3dpca/"+str(gen)
```

```python
ims = []
for i in range(modes):
    im = imread(DIR+'/'+str(i)+'.png')[Hs:He,Ws:We,:]
    ims.append(im)
    
    plt.figure()
    plt.imshow(im)
    plt.grid(False)
    plt.show()
    plt.close()
```

```python
cols  = 4
rows  = int(modes/cols)

SizeH = 1.5
SizeW = 1.5
count = 0

f, axarr = plt.subplots(rows, cols, figsize=((cols)*SizeW,rows*SizeH))

count = 0
for i in range(rows):
    for j in range(cols):
        
        axarr[i,j].imshow(ims[count])    

        axarr[i,j].grid(False)
        axarr[i,j].set_frame_on(False)
        plt.setp(axarr[i,j].get_xticklabels(), visible=False)
        plt.setp(axarr[i,j].get_yticklabels(), visible=False)

        count+= 1

plt.tight_layout()
plt.savefig(DIR+'/plot.png',dpi=500)
plt.savefig(DIR+'/plot.pdf',dpi=500)

plt.show()
plt.close()
```
