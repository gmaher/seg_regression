```python
import numpy as np
import pandas as pd
import matplotlib

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
matplotlib.rc('text', usetex=True)
matplotlib.rcParams.update({"font.size":16})
import os
import sys

sys.path.append(os.path.abspath('../..'))
from modules import io
```

```python
data_dir = "/media/marsdenlab/Data1/UQ"

ylabel = "vWSS std"

quant = 'vWSS_3_boundary'

CSVS = [
data_dir+'/0110/sim_wom_rcr_real/csv/aorta.0.'+quant+'.uncert.csv',
data_dir+'/0110/sim_wom_rcr_real/csv/aorta.1.'+quant+'.uncert.csv',
data_dir+'/0110/sim_wom_rcr_real/csv/right_iliac.1.'+quant+'.uncert.csv',
]

labels = [
    "aorta 0",
    "aorta 1",
    "right iliac 1"
]

colors = [
    "b",
    "r",
    "g"
]

DFS = []
for csv in CSVS:
    df = pd.read_csv(csv)
    DFS.append(df)
```

```python
plt.figure(figsize=(8,8))

for df,l,c in zip(DFS,labels,colors):
    
    plt.plot(df['area_std'], df['val_std'], color=c,label=l, marker='o')
    
plt.xlabel("area std")
plt.ylabel(ylabel)
plt.legend()
# plt.savefig(OUTPUT+'/'+q+'.png',dpi=300)
# plt.savefig(OUTPUT+'/'+q+'.pdf',dpi=200)

plt.show()
plt.close()
```

```python
plt.figure(figsize=(8,8))

for df,l,c in zip(DFS,labels,colors):
    
    plt.plot(df['radius_std'], df['val_std'], color=c,label=l, marker='o')
    
plt.xlabel("radius std")
plt.ylabel(ylabel)
plt.legend()
# plt.savefig(OUTPUT+'/'+q+'.png',dpi=300)
# plt.savefig(OUTPUT+'/'+q+'.pdf',dpi=200)

plt.show()
plt.close()
```
