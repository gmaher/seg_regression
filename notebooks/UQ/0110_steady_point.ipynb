```python
import numpy as np
import pandas as pd
import matplotlib

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
matplotlib.rc('text', usetex=True)
matplotlib.rcParams.update({"font.size":16})
import os
import sys

sys.path.append(os.path.abspath('../..'))
from modules import io
```

```python
CSV = "/media/marsdenlab/Data1/UQ/0110/csv/bif.csv"

df = pd.read_csv(CSV)
print(df.columns)
```

```python
generations = [1,2,3,4]
meshes = ['coarse','medium','fine']
labels = ['pressure_00400_0', 'vWSS_00400_0', 'vWSS_00400_1',
          'vWSS_00400_2', 'velocity_00400_0',
          'velocity_00400_1', 'velocity_00400_2']
colors = ['b','r','r','r','g','g','g']

for g in generations:
    for m in meshes:
        data = df.loc[(df['mesh']==m) & (df['generation']==g)]
        
        for l,c in zip(labels, colors):
            v = data[l]
            
            ll = l.replace('_',' ')
            
            plt.figure()
            plt.title(ll+' '+str(g)+' '+str(m))
            plt.hist(v, color=c, bins=20)
            plt.xlabel(ll)
            plt.ylabel('count')
            plt.show()
```
