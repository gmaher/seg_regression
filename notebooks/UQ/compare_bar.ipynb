```python
import numpy as np
import pandas as pd
import matplotlib

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
matplotlib.rc('text', usetex=True)
matplotlib.rcParams.update({"font.size":16})
import os
import sys

sys.path.append(os.path.abspath('..'))
```

```python
labels = [
    "Coronary pressure waveform",
    "Intramyocardial presure",
    "Morphometry exponent",
    "Vascular wall young's modulus",
    "Boundary conditions",
    "Multiple clinical targets",
    "Full model geometry",
]

p_vals = [
7.0, 0, 7.0, 7.0, 3.0, 0, 0.85
]

p_errs = [
    0, 0, 0, 0, 0, 0, 0.65
]

#TAWSS
t_vals = [
7.0, 25.0, 7.0, 7.0, 15.0, 7.5, 6.5
]

t_errs = [
0,0,0,0,5,2.5,3.5
]

#VELOCITY
v_vals = [
7,25,7,7,1,0,4.25
]

v_errs = [
0,0,0,0,0,0,2.25
]

pos = range(7)

colors = [
  "#6ACC64",
  "#6ACC64",
  "#6ACC64",
  "#6ACC64",

  "#4878D0",
  
  "orange",
    
  "#D65F5F"
]

hatch   = ['', '/', '-', '\\',
           
           '', '', ''
          ]
```

```python
Size  = 6

f, axarr = plt.subplots(1, 1, figsize=(Size,Size))


axarr.errorbar(pos,p_vals,p_errs, marker='s',linestyle='')
axarr.set_ylabel("preesure cv")
axarr.set_xticklabels(labels, rotation="vertical", fontsize=16)

plt.tight_layout()
plt.savefig("uq_compare.pdf",dpi=300)
plt.show()
plt.close()
```

```python
plt.figure(figsize=(6,6))
for l,h,c,v,e,x in zip(labels,hatch,colors,p_vals,p_errs,pos):
    plt.bar(x, height=v, width=0.5, color=c, 
            label=l, yerr=e, hatch=h)

    
lgd = plt.legend(bbox_to_anchor=(0., 1.01, 1., .101), loc=3, ncol=2)

plt.ylabel('pressure cv')
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
plt.ylim(0,27)
#plt.tight_layout()
#plt.savefig('../figures/dice_bar.pdf', bbox_extra_artists=(lgd,), 
#            bbox_inches='tight', pad_inches=0,dpi=500)
plt.grid(True)
plt.xticks([])
plt.show()
plt.close()
```
