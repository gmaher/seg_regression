```python
import numpy as np
import pandas as pd
import matplotlib

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
matplotlib.rc('text', usetex=True)
matplotlib.rcParams.update({"font.size":16})
import os
import sys

sys.path.append(os.path.abspath('../..'))
from modules import io
```

```python
MODELS = [
 "0110",
 "0144",
 "coronary"   
]

PATHS = [
    ["aorta","right_iliac"],
    [
"aorta",   
"celiac_hepatic",
"celiac_splenic",
"ext_iliac_left",
"renal_left",
"renal_right",
"SMA"
],
    ["lc1",
        "lc1_sub1",
        "lc1_sub2",
        "lc1_sub3",
        "lc2",
        "lc2_sub1"]  
]

CASES  = [
"sim_wom_rcr",
"sim_wom_rcr",
"sim_cor_inflow"
]

mesh  = "coarse"

gen   = 2


models   = list(range(500))


for MODEL, PATHS, CASE in zip(MODELS,PATHS,CASES):
    DIR = "/media/marsdenlab/Data1/UQ/"+MODEL+"/generations/"+str(gen)+'/'+mesh

    OUTPUT = "/media/marsdenlab/Data1/UQ/"+MODEL+"/"+CASE+"/csv/"+str(gen)+".csv"

    dfs_tot = []

    for PATH in PATHS:

        dfs = []

        for i in models:
            csv = DIR+'/'+str(i)+'/'+CASE+'/'+PATH+'.csv'
            if not os.path.exists(csv): continue

            dfs.append(pd.read_csv(csv))

        df = pd.concat(dfs)

        df['pressure_0'] = df['pressure_0']/1333

        df['path'] = PATH
        
        dfs_tot.append(df)

    df = pd.concat(dfs_tot)
    df.to_csv(OUTPUT)
```
