```python
import numpy as np
import pandas as pd
import matplotlib

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
matplotlib.rc('text', usetex=True)
matplotlib.rcParams.update({"font.size":16})
import os
import sys

sys.path.append(os.path.abspath('../..'))
from modules import io
```

```python
MESH = 'fine'
CASE_DIR = "/home/marsdenlab/projects/SV/UQ/data/2_vessel_data/data/"+MESH

CSV_FILES = [
 'bifurcation',
 'bifurcation_aorta',
 'bifurcation_right_iliac',
 'inlet',
 'outlet_aorta',
 'outlet_right_iliac'
]

LABELS = [
    "pressure_00400_0",
    "velocity_00400_0",
    "velocity_00400_1",
    "velocity_00400_2",
    "velocity_magnitude"
]

OUT_DIR = "/home/marsdenlab/projects/SV/UQ/data/2_vessel_data/plots/"+MESH

dataframes = []

for fn in CSV_FILES:
    df = pd.read_csv(CASE_DIR+'/'+fn+'.csv')
    
    df['velocity_magnitude'] = (df['velocity_00400_0']**2+\
        df['velocity_00400_1']**2+\
        df['velocity_00400_2']**2)**0.5
    
    dataframes.append(df)
```

```python
for i,fn in enumerate(CSV_FILES):
    df = dataframes[i]
    for l in LABELS:
        xlab = l.replace('_',' ')
        lab = fn.replace('_',' ')
        
        plt.figure()
        plt.hist(df[l],bins=10,label=lab)
        plt.xlabel(xlab)
        plt.legend()
        plt.tight_layout()
        plt.savefig(OUT_DIR+'/'+fn+'_'+l+'.png', dpi=300)
        plt.show()
```
