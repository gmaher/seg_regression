```python
sphere1 = vtk.vtkSphereSource()
sphere1.SetCenter(0,0,0)
sphere1.SetRadius(1.5)
sphere1.Update()

sphere2 = vtk.vtkSphereSource()
sphere2.SetCenter(1.0,0,0)
sphere2.SetRadius(1)
sphere2.Update()

sphere3 = vtk.vtkSphereSource()
sphere3.SetCenter(-1.0,0,0)
sphere3.SetRadius(1)
sphere3.Update()

pds = [sphere1.GetOutput(), sphere2.GetOutput(), sphere3.GetOutput()]
print(pds[0])

output_files = [
    "/home/marsdenlab/projects/SV/UQ/data/0144/coarse/0/sphere1.vtp",
    "/home/marsdenlab/projects/SV/UQ/data/0144/coarse/0/sphere2.vtp",
    "/home/marsdenlab/projects/SV/UQ/data/0144/coarse/0/sphere3.vtp"
]

for pd,o_fn in zip(pds,output_files):
    sv.vtk_write_polydata(pd,o_fn)
```
