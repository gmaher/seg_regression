```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
matplotlib.rc('text', usetex=True)
matplotlib.rcParams.update({"font.size":16})
```

```python
d = 2
K = 40

V = np.random.randn(d,K)
```

```python
plt.figure(figsize=(6,6))
plt.plot(V[0,:],V[1,:],linestyle='',marker='o')
plt.show()
plt.close()
```

```python
alpha = np.random.randn(K)
y = V.dot(alpha)
```

```python
plt.figure(figsize=(6,6))
plt.plot(V[0,:],V[1,:],linestyle='',marker='o')
plt.plot(y[0],y[1],linestyle='',marker='o',color='r')
plt.show()
plt.close()
```

```python
drop  = 0.25
nvec  = 100
Ydrop = np.zeros((d,nvec))

for i in range(nvec):
    z = np.random.binomial(1,1-drop,size=(K))
    alpha_new = alpha*z/(1-drop)
    Ydrop[:,i] = V.dot(alpha_new)
```

```python
plt.figure(figsize=(6,6))
plt.plot(Ydrop[0,:],Ydrop[1,:],linestyle='',marker='o')
plt.plot(y[0],y[1],linestyle='',marker='o',color='r')
plt.ylim(y[1]-40,y[1]+40)
plt.xlim(y[0]-40,y[0]+40)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('dropout p={}'.format(drop))

plt.tight_layout()
plt.savefig('drop_{}.png'.format(drop),dpi=300)
plt.savefig('drop_{}.pdf'.format(drop),dpi=300)

plt.show()
plt.close()
```
