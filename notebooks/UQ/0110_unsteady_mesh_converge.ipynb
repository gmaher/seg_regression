```python
import numpy as np
import pandas as pd
import matplotlib

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
matplotlib.rc('text', usetex=True)
matplotlib.rcParams.update({"font.size":16})
import os
import sys

sys.path.append(os.path.abspath('../..'))
from modules import io
```

```python
paths = [
    "aorta.0",
]

CASES = [
    "sim_wom_rcr_real"
]

meshes = [
    "coarse",
    "medium",
    "fine",
    "fine_2",
    "fine_3"
]

quants   = [
 "pressure_0",
 "vWSS_3_boundary",
 "velocity_3"
]

generation = "converge"

models   = list(range(1))
dt       = 0.02

tups = []

for case in CASES:
    for p in paths:
            for m in meshes:
                    CSV    = "/media/marsdenlab/Data1/UQ/0110/"+case+"/converge"+'/'+p+".csv"
                    
                    OUTPUT = "/media/marsdenlab/Data1/UQ/0110/"+case+"/"+str(generation)+"/"+m+"/"+p+"/tube"
                    io.mkdir(OUTPUT)
                    OUTPUT = OUTPUT+'/'+m
                    io.mkdir(OUTPUT)
                    print(CSV)
                    df = pd.read_csv(CSV)
                    df = df.loc[df['mesh']==m]
                    
                    
                    
                    tups.append( (case,p,m,df,OUTPUT) )
                    
DATA = pd.concat([t[3] for t in tups])
```

```python
d = tups[0][3]
print(d.shape)
print(len(tups))
```

```python
for t in tups:
    case,path,mesh,df,OUTPUT = t
    
    for q in quants:
        series_x = []
        series_y = []
        for m in models:
            data = df.loc[(df['model'] == m) & (df['generation'] == generation)]

            if data.shape[0] == 0: continue

            y = data[q].values
            x = data['time'].values*dt

            series_x.append(x)
            series_y.append(y)

        Y = np.array(series_y)
        X = np.array(series_x)
        
        label = q.split('_')[0]
        
        plt.figure(figsize=(8,8))
        plt.plot(X.T,Y.T, color='b')
        plt.xlabel("time")
        plt.ylabel(label)
        plt.legend()
        plt.savefig(OUTPUT+'/'+q+'.png',dpi=300)
        plt.savefig(OUTPUT+'/'+q+'.pdf',dpi=200)
        
        plt.show()
        plt.close()
```

```python
T = 50
mesh_sizes = [0.5,0.33,0.22,0.15]
for q in quants:
        label = q.replace('_','')
        err = []
        y_fine = DATA.loc[DATA['mesh']=='fine_3'][q].values[-T:]
        
        for mesh in ['coarse','medium','fine','fine_2']:
            y_mesh = DATA.loc[DATA['mesh']==mesh][q].values[-T:]

            err.append(np.amax(np.abs((y_fine-y_mesh)/y_fine)))

        
        plt.figure(figsize=(8,8))
        plt.plot(mesh_sizes, err, label=label)
        plt.xlabel("mesh size")
        plt.ylabel(label)
        plt.legend()
        plt.savefig(OUTPUT+'/'+q+'_err_converge.png',dpi=300)
        plt.savefig(OUTPUT+'/'+q+'_err_converge.pdf',dpi=200)

        plt.show()
        plt.close()
```
