```python
import numpy as np
import pandas as pd
import matplotlib

import os
import sys

sys.path.append(os.path.abspath('..'))
from modules import vascular_data as sv
import vtk
```

```python
pd_fn = "/home/marsdenlab/projects/SV/UQ/data/0144/coarse/0/exterior.vtp"

pd = sv.vtk_read_native_polydata(pd_fn)

print(pd)
```

```python
normals = vtk.vtkPolyDataNormals()
normals.SetInputData(pd)
normals.ComputePointNormalsOn()
normals.ComputeCellNormalsOff()
normals.Update()

pd = normals.GetOutput()
print(pd)
```

```python
points = pd.GetPoints()
N = points.GetNumberOfPoints()
n_data = pd.GetPointData().GetArray("Normals")
```

```python
#0.138737,0.236429,-0.961693
p = [0.0181,14.8262,-35.456299]
for i in range(N):
    point = points.GetPoint(i)
    d = np.sqrt( (point[0]-p[0])**2 + (point[1]-p[1])**2 + (point[2]-p[2])**2 )
    if d < 0.1:
        print(points.GetPoint(i))
        print(n_data.GetTuple(i))
```
